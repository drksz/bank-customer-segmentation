```python
# aggregate transaction amounts from the five dataframes in preparation for merging with cust. info
cc_agg = credit_card.groupby('cust_num').agg(
    total_cc_spend = ('amount','sum'),
    avg_cc_spend = ('amount','mean')
)

dc_agg = debit_card.groupby('cust_num').agg(
    total_dc_spend = ('amount','sum'),
    avg_dc_spend = ('amount','mean')
)

dt_agg = digital_transact.groupby('cust_num').agg(
    total_dig_transact = ('amount','sum'),
    avg_dig_transact = ('amount','mean')
)

insta_in_agg = instapay_incoming.groupby('cust_num').agg(
    total_instapay_in = ('amount','sum'),
    avg_instapay_in = ('amount','mean')
)

insta_out_agg = instapay_outgoing.groupby('cust_num').agg(
    total_instapay_out = ('amount','sum'),
    avg_instapay_out = ('amount','mean')
)
```

```python
cc_agg.head()
```

```python
merge1 = cust_info.merge(cc_agg, how='left', on='cust_num')
merge2 = merge1.merge(dc_agg, how='left', on='cust_num')
merge3 = merge2.merge(dt_agg, how='left',on='cust_num')
merge4 = merge3.merge(insta_in_agg, how='left', on='cust_num')
cust_info_main = merge4.merge(insta_out_agg, how='left', on='cust_num')
```

```python
cust_info_main = cust_info_main.drop(columns=['segment','subsegment','idv_or_org'])
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
cust_info_main.columns
```

```python

feature_set = ['total_cc_spend','total_dc_spend','total_instapay_in','total_instapay_out','total_dig_transact']

cust_info_main[feature_set] = cust_info_main[feature_set].fillna(0)

cust_features = cust_info_main[feature_set]
```

```python
scaler = preprocessing.StandardScaler()
scaled_cust_info = scaler.fit_transform(cust_features)
```

```python
pca = PCA()

pca.fit(scaled_cust_info)
pca_data = pca.transform(scaled_cust_info)
```

```python
# calculate percentage of variation each PC accounts for 
pct_var = np.round(pca.explained_variance_ratio_ * 100, decimals=1)

labels = ['PC' + str(x) for x in range(1, len(pct_var) + 1)]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

ax.bar(x=range(1, len(pct_var) + 1), height=pct_var)
ax.set_xticklabels(labels)

ax.set_ylabel('Percentage of explained variance')
ax.set_xlabel('Principal component')
plt.title('Scree plot')

plt.show()
```

```python
pca_df = pd.DataFrame(pca_data, columns=labels, index=cust_features.index)
```

```python
pca_df.head()
```

```python
pca_df['digital_flag'] = cust_info_main['digital_flag']

fig, ax = plt.subplots(1, 1, figsize=(10, 8))

sns.scatterplot(
    data=pca_df, 
    x='PC2',
    y='PC3',
    alpha=0.6,
    hue='digital_flag',
    ax=ax
)

ax.set_title('PCA scatterplot: PC2 vs PC3')

plt.show()
```

```python
loadings = pd.DataFrame(
    pca.components_.T, 
    columns=labels,             # ['PC1', 'PC2', 'PC3', 'PC4', 'PC5']
    index=cust_features.columns # original feature names
)
```

```python
loadings.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

sns.scatterplot(data=pca_df, x='PC2', y='PC3', color='grey', alpha=0.4, ax=ax)

# plot the features/loadings as arrows and annotations
# max value of PC2 is used to scale the arrows so they are visible over the scatterplot
scale_factor = pca_df['PC2'].max() 

for feature in loadings.index:
    # get the X (PC2) and Y (PC3) coordinates for the feature's arrow
    x_val = loadings.loc[feature, 'PC2'] * scale_factor
    y_val = loadings.loc[feature, 'PC3'] * scale_factor
    
    # draw the arrow starting from the origin (0,0)
    ax.arrow(0, 0, x_val, y_val, color='red', alpha=0.8, width=0.02, head_width=0.15)
    
    # annotate the column name right at the tip of the arrow
    ax.annotate(feature, (x_val, y_val * 1.05), color='maroon', fontsize=12, weight='bold')

# clean up the plot and label the axes properly
ax.set_title('PCA Biplot: Customers and Feature Loadings (PC2 vs PC3)')
ax.set_xlabel('Principal Component 2 (PC2)')
ax.set_ylabel('Principal Component 3 (PC3)')
ax.axhline(0, color='black', linestyle='--', alpha=0.3) # add X-axis line at 0
ax.axvline(0, color='black', linestyle='--', alpha=0.3) # add Y-axis line at 0

plt.show()
```
