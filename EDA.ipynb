```python
# aggregate transaction amounts from the five dataframes in preparation for merging with cust. info
cc_agg = credit_card.groupby('cust_num').agg(
    total_cc_spend = ('amount','sum'),
    avg_cc_spend = ('amount','mean')
)

dc_agg = debit_card.groupby('cust_num').agg(
    total_dc_spend = ('amount','sum'),
    avg_dc_spend = ('amount','mean')
)

dt_agg = digital_transact.groupby('cust_num').agg(
    total_dig_transact = ('amount','sum'),
    avg_dig_transact = ('amount','mean')
)

insta_in_agg = instapay_incoming.groupby('cust_num').agg(
    total_instapay_in = ('amount','sum'),
    avg_instapay_in = ('amount','mean')
)

insta_out_agg = instapay_outgoing.groupby('cust_num').agg(
    total_instapay_out = ('amount','sum'),
    avg_instapay_out = ('amount','mean')
)
```

```python
cc_agg.head()
```

```python
merge1 = cust_info.merge(cc_agg, how='left', on='cust_num')
merge2 = merge1.merge(dc_agg, how='left', on='cust_num')
merge3 = merge2.merge(dt_agg, how='left',on='cust_num')
merge4 = merge3.merge(insta_in_agg, how='left', on='cust_num')
cust_info_main = merge4.merge(insta_out_agg, how='left', on='cust_num')
```

```python
cust_info_main = cust_info_main.drop(columns=['segment','subsegment','idv_or_org'])
```

```python
cust_info_main.head()
```

```python
has_cc_txn = cust_info_main.loc[cust_info_main['total_cc_spend'].notna(), :]
has_dc_txn = cust_info_main.loc[cust_info_main['total_dc_spend'].notna(), :]
has_digital_txn = cust_info_main.loc[cust_info_main['total_dig_transact'].notna(), :]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 5), sharey=True)
sns.histplot(has_cc_txn['total_cc_spend'], ax=ax[0], bins=20)
ax[0].set_title('Total credit card spend distribution')

sns.histplot(has_dc_txn['total_dc_spend'], ax=ax[1], bins=20)
ax[1].set_title('Total debit card spend distribution')

sns.histplot(has_digital_txn['total_dig_transact'], ax=ax[2], bins=20)
ax[2].set_title('Total digital transaction spend distribution')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(3, 2, figsize=(12, 15))

sns.scatterplot(
    data=cust_info_main,
    x='age',
    y='total_cc_spend',
    hue='business_owner',
    ax=ax[0, 0],
    alpha=0.5
)

sns.scatterplot(
    data=cust_info_main,
    x='age',
    y='total_dc_spend',
    hue='business_owner',
    ax=ax[0, 1],
    alpha=0.5
)

sns.scatterplot(
    data=cust_info_main,
    x='age',
    y='total_dig_transact',
    hue='business_owner',
    ax=ax[1, 0],
    alpha=0.5
)

sns.scatterplot(
    data=cust_info_main,
    x='age',
    y='total_instapay_in',
    hue='business_owner',
    ax=ax[1, 1],
    alpha=0.5
)

sns.scatterplot(
    data=cust_info_main,
    x='age',
    y='total_instapay_out',
    hue='business_owner',
    ax=ax[2, 0],
    alpha=0.5
)

fig.delaxes(ax[2, 1])

plt.show()
```
