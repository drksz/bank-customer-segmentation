```python
# aggregate transaction amounts from the five dataframes in preparation for merging with cust. info
cc_agg = credit_card.groupby('cust_num').agg(
    total_cc_spend = ('amount','sum'),
    avg_cc_spend = ('amount','mean')
)

dc_agg = debit_card.groupby('cust_num').agg(
    total_dc_spend = ('amount','sum'),
    avg_dc_spend = ('amount','mean')
)

dt_agg = digital_transact.groupby('cust_num').agg(
    total_dig_transact = ('amount','sum'),
    avg_dig_transact = ('amount','mean')
)

insta_in_agg = instapay_incoming.groupby('cust_num').agg(
    total_instapay_in = ('amount','sum'),
    avg_instapay_in = ('amount','mean')
)

insta_out_agg = instapay_outgoing.groupby('cust_num').agg(
    total_instapay_out = ('amount','sum'),
    avg_instapay_out = ('amount','mean')
)
```

```python
merge1 = cust_info.merge(cc_agg, how='left', on='cust_num')
merge2 = merge1.merge(dc_agg, how='left', on='cust_num')
merge3 = merge2.merge(dt_agg, how='left',on='cust_num')
merge4 = merge3.merge(insta_in_agg, how='left', on='cust_num')
cust_info_main = merge4.merge(insta_out_agg, how='left', on='cust_num')
```

```python
cc_last_txn = (
    credit_card
    .groupby('cust_num')['date']
    .max()
    .reset_index(name='last_txn_date')
)

dc_last_txn = (
    debit_card
    .groupby('cust_num')['date']
    .max()
    .reset_index(name='last_txn_date')
)

digital_last_txn = (
    digital_transact
    .groupby('cust_num')['date']
    .max()
    .reset_index(name='last_txn_date')
)

# reference date is the endpoint date of the collected data
reference_date = pd.to_datetime('2023-06-30')


cc_last_txn['days_last_cc_txn'] = (
    reference_date - pd.to_datetime(cc_last_txn['last_txn_date'])
).dt.days

dc_last_txn['days_last_dc_txn'] = (
    reference_date - pd.to_datetime(dc_last_txn['last_txn_date'])
).dt.days

digital_last_txn['days_last_dig_txn'] = (
    reference_date - pd.to_datetime(digital_last_txn['last_txn_date'])
).dt.days
```

```python
cust_info_main = cust_info_main.merge(
    cc_last_txn[['cust_num','days_last_cc_txn']],
    on='cust_num',
    how='left'
)

cust_info_main = cust_info_main.merge(
    dc_last_txn[['cust_num','days_last_dc_txn']],
    on='cust_num',
    how='left'
)

cust_info_main = cust_info_main.merge(
    digital_last_txn[['cust_num','days_last_dig_txn']],
    on='cust_num',
    how='left'
)

# fill null values with 91 to indicate that the customer has been inactive for a while 
cust_info_main['days_last_cc_txn'] = cust_info_main['days_last_cc_txn'].fillna(91).astype(int)
cust_info_main['days_last_dc_txn'] = cust_info_main['days_last_dc_txn'].fillna(91).astype(int)
cust_info_main['days_last_dig_txn'] = cust_info_main['days_last_dig_txn'].fillna(91).astype(int)
```

```python
cust_info_main = cust_info_main.drop(columns=['segment','subsegment','idv_or_org'])
```

```python
fill_cols = ['total_cc_spend', 'total_dc_spend', 'total_dig_transact', 'total_instapay_in', 'total_instapay_out'] 

cust_info_main[fill_cols] = cust_info_main[fill_cols].fillna(0)
```

```python
total_expenditure = ( cust_info_main['total_cc_spend'] 
                     + cust_info_main['total_dc_spend'] 
                     + cust_info_main['total_dig_transact'] 
)
```

```python
cust_info_main['cc_share'] = cust_info_main['total_cc_spend'] / total_expenditure
cust_info_main['dc_share'] = cust_info_main['total_dc_spend'] / total_expenditure
cust_info_main['digital_share'] = cust_info_main['total_dig_transact'] / total_expenditure

share_cols = ['cc_share', 'dc_share', 'digital_share']
cust_info_main[share_cols] = cust_info_main[share_cols].fillna(0)
```

```python
outflow = (
    cust_info_main['total_cc_spend'] + 
    cust_info_main['total_dc_spend'] + 
    cust_info_main['total_dig_transact'] + 
    cust_info_main['total_instapay_out']
)

cust_info_main['total_outflow'] = outflow
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
feature_set = [
    'total_cc_spend',
    'total_dc_spend',
    'total_instapay_in',
    'total_instapay_out',
    'total_dig_transact',
    'age',
    'tenure',
    'days_last_cc_txn',
    'days_last_dc_txn',
    'days_last_dig_txn', 
    'cc_share',
    'dc_share',
    'digital_share',
    'total_outflow'
]
```

```python
# Another sanity check NaN fill
cust_features = cust_info_main.loc[:, feature_set].fillna(0)
```

```python
#cust_features = np.log1p(cust_features)
scaler = preprocessing.StandardScaler()
scaled_cust_info = scaler.fit_transform(cust_features)
```

```python
pca = PCA()

pca.fit(scaled_cust_info)
pca_data = pca.transform(scaled_cust_info)
```

```python
# calculate percentage of variation each PC accounts for 
pct_var = np.round(pca.explained_variance_ratio_ * 100, decimals=1)

labels = ['PC' + str(x) for x in range(1, len(pct_var) + 1)]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 5))

ax.bar(x=range(1, len(pct_var) + 1), height=pct_var)
ax.set_xticklabels(labels)

ax.set_ylabel('Percentage of explained variance')
ax.set_xlabel('Principal component')
plt.title('Scree plot')

plt.show()
```

```python
pca_df = pd.DataFrame(pca_data, columns=labels, index=cust_features.index)
```

```python
pca_df.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 5))

sns.scatterplot(
    data=pca_df, 
    x='PC2',
    y='PC3',
    alpha=0.6,
    ax=ax
)

ax.set_title('PCA scatterplot: PC2 vs PC3')

plt.show()
```

```python
loadings = pd.DataFrame(
    pca.components_.T, 
    columns=labels,             # ['PC1', 'PC2', 'PC3', 'PC4', 'PC5']
    index=cust_features.columns # original feature names
)
```

```python
pca_df_plot = pca_df.copy()
pca_df_plot['digital_flag'] = cust_info_main['digital_flag']


fig, ax = plt.subplots(1, 1, figsize=(12, 8))

sns.scatterplot(data=pca_df_plot, x='PC2', y='PC3', color='grey', alpha=0.3, ax=ax, hue='digital_flag')

# plot the features/loadings as arrows and annotations
# max value of PC2 is used to scale the arrows so they are visible over the scatterplot
scale_factor = pca_df['PC2'].max() 

for feature in loadings.index:
    # get the X (PC2) and Y (PC3) coordinates for the feature's arrow
    x_val = loadings.loc[feature, 'PC2'] * scale_factor
    y_val = loadings.loc[feature, 'PC3'] * scale_factor
    
    ax.arrow(0, 0, x_val, y_val, color='red', alpha=0.8, width=0.02, head_width=0.15)
    ax.annotate(feature, (x_val, y_val * 1.05), color='black', fontsize=10)

# clean up the plot and label the axes properly
ax.set_title('PCA Biplot: Customers and Feature Loadings (PC2 vs PC3)')
ax.set_xlabel('Principal Component 2 (PC2)')
ax.set_ylabel('Principal Component 3 (PC3)')
ax.axhline(0, color='black', linestyle='--', alpha=0.3) # add X-axis line at 0
ax.axvline(0, color='black', linestyle='--', alpha=0.3) # add Y-axis line at 0

plt.show()
```

```python
# heuristic tagging using PC2 and PC3 loading scores.
# these principal components will represent a scoring index,
# 'traditional spending score' and 'digital outflow score' respectively.
# this approach aims to categorize customers using rule-based segmentation.

cust_info_main['traditional_spend_score'] = pca_df['PC2']
cust_info_main['digital_outflow_score'] = pca_df['PC3']

print(cust_info_main[['cust_num','digital_outflow_score','traditional_spend_score']].head())
```

```python
# the method below will be applied to each row in the cust_info_main
# table in a vectorized fashion. Three rules are established

def categorize_cust(row):

    # a row/cust record is automatically categorized as a 'CC User',
    # knowing that CC users only make up less than 5% of the entire data. 
    # this rule protects that % metric
    if row['cc_share'] > 0:
        return 'Long-term CC Users'
    
    # a below zero score for digital outflow marks a user as a cash mover. 
    # this mirrors the lower right cluster in the biplot.
    elif row['digital_outflow_score'] < 0:
        return 'Cash Movers'
    
    # a default rule that categorizes a user as a debit card loyalist
    # if cc share is zero or negative and outflow score is high
    else:
        return 'Loyal Debit Card Users'
```

```python
# apply the function to the rows of the df
cust_info_main['cust_segment'] = cust_info_main.apply(categorize_cust, axis=1)
```
